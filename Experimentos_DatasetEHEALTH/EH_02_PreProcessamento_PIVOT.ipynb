### importando

```python
## importando dataset outputPx.csv
df = pd.read_csv('dataseteh/pp2/eh_pp2.csv',delimiter=';',decimal=',',thousands=".",encoding='ANSI')
print(df.shape)
df.head()
```

##  Pivoteamento  dia e nome da atividade e FREQUENCIA

```python
## pivotanto usando dia e nome da atividade e Frequencia
dfpiv_freq = df.pivot_table(values = 'frequencia', 
                         index = ['dia','dia_semana','fim_semana'], 
                         columns = 'atividade')
dfpiv_freq.fillna(value=0, inplace=True)
dfpiv_freq = dfpiv_freq.reset_index()
dfpiv_freq.head()
```

```python
dfpiv_freq.rename(columns = {'Change clothes':'cclothes','Eating':'eating','Make coffee':'mcoffe',
                     'Make hot food':'mhotfood','Toileting':'toileting','Washing hand/face':'whandface'}, inplace = True)
print(dfpiv_freq.shape)
dfpiv_freq.head()
```

##  Pivoteamento  dia e nome da atividade e DURACAO

```python
## pivotanto usando dia e nome da atividade e DURACAO
dfpiv_duracao = df.pivot_table(values = 'duracao', 
                         index = ['dia','dia_semana','fim_semana'], 
                         columns = 'atividade')
dfpiv_duracao.fillna(value=0, inplace=True)
dfpiv_duracao = dfpiv_duracao.reset_index()
dfpiv_duracao.head()
```

```python
dfpiv_duracao.rename(columns = {'Change clothes':'cclothes','Eating':'eating','Make coffee':'mcoffe',
                     'Make hot food':'mhotfood','Toileting':'toileting','Washing hand/face':'whandface'}, inplace = True)
print(dfpiv_duracao.shape)
dfpiv_duracao.head()
```

##  Pivoteamento  dia e nome da atividade e DURACAO/FREQUENCIA

```python
## pivotanto usando dia e nome da atividade e DURACAO
dfpiv_dfreq = df.pivot_table(values = 'durfreq', 
                         index = ['dia','dia_semana','fim_semana'], 
                         columns = 'atividade')
dfpiv_dfreq.fillna(value=0, inplace=True)
dfpiv_dfreq = dfpiv_dfreq.reset_index()
dfpiv_dfreq.head()
```

```python
dfpiv_dfreq.rename(columns = {'Change clothes':'cclothes','Eating':'eating','Make coffee':'mcoffe',
                     'Make hot food':'mhotfood','Toileting':'toileting','Washing hand/face':'whandface'}, inplace = True)
print(dfpiv_dfreq.shape)
dfpiv_dfreq.head()
```

```python
#exportando os dataset
dfpiv_duracao.to_csv(r'dataseteh/pp3/eh-dur_pp3.csv', sep=';', index=False, decimal=',')
```

```python
#exportando os dataset
dfpiv_dfreq.to_csv(r'dataseteh/pp3/eh-dfreq_pp3.csv', sep=';', index=False, decimal=',')
```
