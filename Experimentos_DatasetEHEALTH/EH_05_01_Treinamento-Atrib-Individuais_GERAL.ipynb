## Treinando MODELOS (4) para SEMANAS (18) para DURACAO

```python
# inicializando listas para armazenar as métricas

for k in range (1,6): #iterando variacoes
    for z in range (1,4): #iterando atributos
        for y in range (1,5): #iteracao modelos

            csemana = []
            cvariacao = []
            cvp = []
            cvn = []
            cfp = []
            cfn = []
            cacuracia = []
            cprecisao = []
            csensibilidade = []
            cf1escore = []

            for x in range(1,19): #iterando semanas
                semana = x
                # ATRIBUTO DURACAO
                if (z == 1) & (k == 1): #duracao 20-35
                    df_test  = df_dur_test2035.iloc[:, [3,4,5,6,7]].copy()
                    df_dur_test = df_dur_test2035.copy()
                    df_trein = df_dur_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'duracao'
                    variacao = '20-35'
                if (z == 1) & (k == 2): #duracao 40-55
                    df_test  = df_dur_test4055.iloc[:, [3,4,5,6,7]].copy()
                    df_dur_test = df_dur_test4055.copy()
                    df_trein = df_dur_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'duracao'
                    variacao = '40-55'
                if (z == 1) & (k == 3): #duracao 60-75
                    df_test  = df_dur_test6075.iloc[:, [3,4,5,6,7]].copy()
                    df_dur_test = df_dur_test6075.copy()
                    df_trein = df_dur_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'duracao'
                    variacao = '60-75'
                if (z == 1) & (k == 4): #duracao 80-95
                    df_test  = df_dur_test8095.iloc[:, [3,4,5,6,7]].copy()
                    df_dur_test = df_dur_test8095.copy()
                    df_trein = df_dur_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'duracao'
                    variacao = '80-95'      
                if (z == 1) & (k == 5): #duracao 20-95
                    df_test  = df_dur_test2095.iloc[:, [3,4,5,6,7]].copy()
                    df_dur_test = df_dur_test2095.copy()
                    df_trein = df_dur_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'duracao'
                    variacao = '20-95'                    
                    
                # ATRIBUTO FREQUENCIA
                if (z == 2) & (k == 1): #freq 20-35
                    df_test  = df_freq_test2035.iloc[:, [3,4,5,6,7]].copy()
                    df_freq_test = df_freq_test2035.copy()
                    df_trein = df_freq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'frequencia'
                    variacao = '20-35'
                if (z == 2) & (k == 2): #duracao 40-55
                    df_test  = df_freq_test4055.iloc[:, [3,4,5,6,7]].copy()
                    df_freq_test = df_freq_test4055.copy()
                    df_trein = df_freq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'frequencia'
                    variacao = '40-55'
                if (z == 2) & (k == 3): #duracao 60-75
                    df_test  = df_freq_test6075.iloc[:, [3,4,5,6,7]].copy()
                    df_freq_test = df_freq_test6075.copy()
                    df_trein = df_freq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'frequencia'
                    variacao = '60-75'
                if (z == 2) & (k == 4): #duracao 80-95
                    df_test  = df_freq_test8095.iloc[:, [3,4,5,6,7]].copy()
                    df_freq_test = df_freq_test8095.copy()
                    df_trein = df_freq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'frequencia'
                    variacao = '80-95'      
                if (z == 2) & (k == 5): #duracao 20-95
                    df_test  = df_freq_test2095.iloc[:, [3,4,5,6,7]].copy()
                    df_freq_test = df_freq_test2095.copy()
                    df_trein = df_freq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'frequencia'
                    variacao = '20-95'                     
                    
                # ATRIBUTO DFREQ
                if (z == 3) & (k == 1): #dfreq 20-35
                    df_test  = df_dfreq_test2035.iloc[:, [3,4,5,6,7]].copy()
                    df_dfreq_test = df_dfreq_test2035.copy()
                    df_trein = df_dfreq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'dfreq'
                    variacao = '20-35'
                if (z == 3) & (k == 2): #dfreq 40-55
                    df_test  = df_dfreq_test4055.iloc[:, [3,4,5,6,7]].copy()
                    df_dfreq_test = df_dfreq_test4055.copy()
                    df_trein = df_dfreq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'dfreq'
                    variacao = '40-55'
                if (z == 3) & (k == 3): #duracao 60-75
                    df_test  = df_dfreq_test6075.iloc[:, [3,4,5,6,7]].copy()
                    df_dfreq_test = df_dfreq_test6075.copy()
                    df_trein = df_dfreq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'dfreq'
                    variacao = '60-75'
                if (z == 3) & (k == 4): #duracao 80-95
                    df_test  = df_dfreq_test8095.iloc[:, [3,4,5,6,7]].copy()
                    df_dfreq_test = df_dfreq_test8095.copy()
                    df_trein = df_dfreq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'dfreq'
                    variacao = '80-95'   
                if (z == 3) & (k == 5): #duracao 20-95
                    df_test  = df_dfreq_test2095.iloc[:, [3,4,5,6,7]].copy()
                    df_dfreq_test = df_dfreq_test2095.copy()
                    df_trein = df_dfreq_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5,6,7]]
                    atributo = 'dfreq'
                    variacao = '20-95'                     
                    
                #standarização
                scaler = StandardScaler().fit(df_trein)
                df_trein = scaler.transform(df_trein).copy()
                df_test = scaler.transform(df_test).copy()

                # aplicando o primeiro algoritmo o OSMV
                # fit the model
                if y == 1:
                    clf = svm.OneClassSVM(nu=0.1, kernel='rbf', gamma='scale')
                    algoritmo = 'ocsvm'
                if y == 2:
                    clf = LocalOutlierFactor(n_neighbors=10, novelty=True)
                    algoritmo = 'lof'
                if y == 3:
                    clf = IsolationForest(contamination= 0.01, random_state=1)
                    algoritmo = 'iforest'
                if y == 4:
                    clf = EllipticEnvelope(contamination= 0.01, random_state=0, support_fraction= 0.9)
                    algoritmo = 'rce'

                clf.fit(df_trein)
                y_pred_train = clf.predict(df_trein)
                y_pred_teste = clf.predict(df_test)
                n_error_train = y_pred_train[y_pred_train == -1].size

                # calculando métricas
                if z == 1 : #duracao
                    df_dur_test['predicao'] = y_pred_teste
                    df_dur_test['vp'] = np.where((df_dur_test['outlier'] == -1) & (df_dur_test['predicao'] == -1), 1,0)
                    df_dur_test['vn'] = np.where((df_dur_test['outlier'] == 1) & (df_dur_test['predicao'] == 1), 1,0)
                    df_dur_test['fp'] = np.where((df_dur_test['outlier'] == -1) & (df_dur_test['predicao'] == 1), 1,0)
                    df_dur_test['fn'] = np.where((df_dur_test['outlier'] == 1) & (df_dur_test['predicao'] == -1), 1,0)
                    VP = df_dur_test['vp'].sum()
                    VN = df_dur_test['vn'].sum()
                    FP = df_dur_test['fp'].sum()
                    FN = df_dur_test['fn'].sum()
                if z == 2 : #freq
                    df_freq_test['predicao'] = y_pred_teste
                    df_freq_test['vp'] = np.where((df_freq_test['outlier'] == -1) & (df_freq_test['predicao'] == -1), 1,0)
                    df_freq_test['vn'] = np.where((df_freq_test['outlier'] == 1) & (df_freq_test['predicao'] == 1), 1,0)
                    df_freq_test['fp'] = np.where((df_freq_test['outlier'] == -1) & (df_freq_test['predicao'] == 1), 1,0)
                    df_freq_test['fn'] = np.where((df_freq_test['outlier'] == 1) & (df_freq_test['predicao'] == -1), 1,0)
                    VP = df_freq_test['vp'].sum()
                    VN = df_freq_test['vn'].sum()
                    FP = df_freq_test['fp'].sum()
                    FN = df_freq_test['fn'].sum()                
                if z == 3 : #dfreq
                    df_dfreq_test['predicao'] = y_pred_teste
                    df_dfreq_test['vp'] = np.where((df_dfreq_test['outlier'] == -1) & (df_dfreq_test['predicao'] == -1), 1,0)
                    df_dfreq_test['vn'] = np.where((df_dfreq_test['outlier'] == 1) & (df_dfreq_test['predicao'] == 1), 1,0)
                    df_dfreq_test['fp'] = np.where((df_dfreq_test['outlier'] == -1) & (df_dfreq_test['predicao'] == 1), 1,0)
                    df_dfreq_test['fn'] = np.where((df_dfreq_test['outlier'] == 1) & (df_dfreq_test['predicao'] == -1), 1,0)
                    VP = df_dfreq_test['vp'].sum()
                    VN = df_dfreq_test['vn'].sum()
                    FP = df_dfreq_test['fp'].sum()
                    FN = df_dfreq_test['fn'].sum()                

                erro = (FP+FN)/(VP+VN+FP+FN) #erro
                acuracia =  (VP+VN)/(VP+VN+FP+FN) # accuracy
                precisao = (VP)/(FP+VP) # precision
                sensibilidade = VP/(FN+VP) #recall
                f1escore = 2 * ((precisao * sensibilidade)/(precisao + sensibilidade))
                #armazenando metricas temp
                csemana.append(semana)
                cvariacao.append(variacao)
                cvp.append(VP)
                cvn.append(VN)
                cfp.append(FP)
                cfn.append(FN)
                cacuracia.append(acuracia)
                cprecisao.append(precisao)
                csensibilidade.append(sensibilidade)
                cf1escore.append(f1escore)
                # dentro do for X SEMANAS

            #criando um dataframe a partir das métricas    
            zipped = list(zip(csemana,cvariacao,cvp,cvn,cfp,cfn,cacuracia, cprecisao, csensibilidade, cf1escore))    
            dfmetricas = pd.DataFrame(zipped, columns=['semanas','variacao','vp', 'vn', 'fp', 'fn','acuracia','precisao','sensibilidade','f1score'])
            dfmetricas['algoritmo'] = algoritmo
            dfmetricas['variavel'] = atributo
            dfmetricas['variacao'] = variacao
            dfmetricas= dfmetricas.replace(np.nan,0)
            #print(dfmetricas.head(20))
            #exportando dataframes para datasets

            dfmetricas.to_csv(r'dataseteh/results/armetricas-'+algoritmo+'-'+atributo+'-var-'+variacao+'-trein126.csv', sep=';', index=False, decimal=',')

            #dentro do for Y MODELOS
        #dentro do for Z ATRIBUTOS
```
