```python
dfc_metricas = pd.DataFrame()

for k in range (1,6): #ITERANDO VARIACOES
    
    for z in range (1,6): #ITERANDO ATIVIDADES

        for y in range (1,5): #ITERANDO ALGORITMOS    

            
            for semana in range(1,9): #iterando semanas
                cont_semana = str(semana)+''
                
                # inicializando listas para armazenar as métricas
                cvp = []
                cvn = []
                cfp = []
                cfn = []
                cacuracia = []
                cprecisao = []
                csensibilidade = []
                cf1escore = []
                cvariavel = []
                csemana = []

                #BedToToilet
                if (z==1) & (k==1): #BedToToilet 20-35
                    df_test = df_BedToToilet_test2035.iloc[:, [3,4,5]].copy()
                    df_trein = df_BedToToilet_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_BedToToilet_test = df_BedToToilet_test2035.copy()
                    variacao = '20-35'
                    atividade = 'BedToToilet'
                if (z==1) & (k==2): #BedToToilet 40-55
                    df_test = df_BedToToilet_test4055.iloc[:, [3,4,5]].copy()
                    df_trein = df_BedToToilet_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_BedToToilet_test = df_BedToToilet_test4055.copy()
                    variacao = '40-55'
                    atividade = 'BedToToilet'
                if (z==1) & (k==3): #BedToToilet 60-75
                    df_test = df_BedToToilet_test6075.iloc[:, [3,4,5]].copy()
                    df_trein = df_BedToToilet_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()                   
                    df_BedToToilet_test = df_BedToToilet_test6075.copy()
                    variacao = '60-75'
                    atividade = 'BedToToilet'
                if (z==1) & (k==4): #BedToToilet 80-95
                    df_test = df_BedToToilet_test8095.iloc[:, [3,4,5]].copy()
                    df_trein = df_BedToToilet_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()                   
                    df_BedToToilet_test = df_BedToToilet_test8095.copy()
                    variacao = '80-95'
                    atividade = 'BedToToilet'
                if (z==1) & (k==5): #BedToToilet 20-95
                    df_test = df_BedToToilet_test2095.iloc[:, [3,4,5]].copy()
                    df_trein = df_BedToToilet_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()                   
                    df_BedToToilet_test = df_BedToToilet_test2095.copy()
                    variacao = '20-95'
                    atividade = 'BedToToilet'                    

                #Eating    
                if (z==2) & (k==1): #Eating 20-35
                    df_test = df_Eating_test2035.iloc[:, [3,4,5]].copy()
                    df_trein = df_Eating_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()                   
                    df_Eating_test = df_Eating_test2035.copy()
                    variacao = '20-35'
                    atividade = 'Eating'
                if (z==2) & (k==2): #Eating 40-55
                    df_test = df_Eating_test4055.iloc[:, [3,4,5]].copy()
                    df_trein = df_Eating_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_Eating_test = df_Eating_test4055.copy()
                    variacao = '40-55'
                    atividade = 'Eating'
                if (z==2) & (k==3): #Eating 60-75
                    df_test = df_Eating_test6075.iloc[:, [3,4,5]].copy()
                    df_trein = df_Eating_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_Eating_test = df_Eating_test6075.copy()
                    variacao = '60-75'
                    atividade = 'Eating'
                if (z==2) & (k==4): #Eating 80-95
                    df_test = df_Eating_test8095.iloc[:, [3,4,5]].copy()
                    df_trein = df_Eating_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_Eating_test = df_Eating_test8095.copy()
                    variacao = '80-95'
                    atividade = 'Eating'        
                if (z==2) & (k==5): #Eating 20-95
                    df_test = df_Eating_test2095.iloc[:, [3,4,5]].copy()
                    df_trein = df_Eating_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_Eating_test = df_Eating_test2095.copy()
                    variacao = '20-95'
                    atividade = 'Eating'                       

                #MealPreparation    
                if (z==3) & (k==1): #MealPreparation 20-35
                    df_test = df_MealPreparation_test2035.iloc[:, [3,4,5]].copy()
                    df_trein = df_MealPreparation_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_MealPreparation_test = df_MealPreparation_test2035.copy()
                    variacao = '20-35'
                    atividade = 'MealPreparation'
                if (z==3) & (k==2): #MealPreparation 40-55
                    df_test = df_MealPreparation_test4055.iloc[:, [3,4,5]].copy()
                    df_trein = df_MealPreparation_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_MealPreparation_test = df_MealPreparation_test4055.copy()
                    variacao = '40-55'
                    atividade = 'MealPreparation'
                if (z==3) & (k==3): #MealPreparation 60-75
                    df_test = df_MealPreparation_test6075.iloc[:, [3,4,5]].copy()
                    df_trein = df_MealPreparation_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_MealPreparation_test = df_MealPreparation_test6075.copy()
                    variacao = '60-75'
                    atividade = 'MealPreparation'
                if (z==3) & (k==4): #MealPreparation 80-95
                    df_test = df_MealPreparation_test8095.iloc[:, [3,4,5]].copy()
                    df_trein = df_MealPreparation_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_MealPreparation_test = df_MealPreparation_test8095.copy()
                    variacao = '80-95'
                    atividade = 'MealPreparation'      
                if (z==3) & (k==5): #MealPreparation 20-95
                    df_test = df_MealPreparation_test2095.iloc[:, [3,4,5]].copy()
                    df_trein = df_MealPreparation_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_MealPreparation_test = df_MealPreparation_test2095.copy()
                    variacao = '20-95'
                    atividade = 'MealPreparation'                        

                #Relax    
                if (z==4) & (k==1): #Relax 20-35
                    df_test = df_Relax_test2035.iloc[:, [3,4,5]].copy()
                    df_trein = df_Relax_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_SpareTimeTV_test = df_Relax_test2035.copy()
                    variacao = '20-35'
                    atividade = 'Relax'
                if (z==4) & (k==2): #Relax 40-55
                    df_test = df_Relax_test4055.iloc[:, [3,4,5]].copy()
                    df_trein = df_Relax_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_SpareTimeTV_test = df_Relax_test4055.copy()
                    variacao = '40-55'
                    atividade = 'Relax'
                if (z==4) & (k==3): #Relax 60-75
                    df_test = df_Relax_test6075.iloc[:, [3,4,5]].copy()
                    df_trein = df_Relax_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_SpareTimeTV_test = df_Relax_test6075.copy()
                    variacao = '60-75'
                    atividade = 'Relax'
                if (z==4) & (k==4): #Relax 80-95
                    df_test = df_Relax_test8095.iloc[:, [3,4,5]].copy()
                    df_trein = df_Relax_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_SpareTimeTV_test = df_Relax_test8095.copy()
                    variacao = '80-95'
                    atividade = 'Relax'          
                if (z==4) & (k==5): #Relax 20-95
                    df_test = df_Relax_test2095.iloc[:, [3,4,5]].copy()
                    df_trein = df_Relax_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_SpareTimeTV_test = df_Relax_test2095.copy()
                    variacao = '20-95'
                    atividade = 'Relax'                      

                #Sleeping    
                if (z==5) & (k==1): #Sleeping 20-35
                    df_test = df_Sleeping_test2035.iloc[:, [3,4,5]].copy()
                    df_trein = df_Sleeping_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_Sleeping_test = df_Sleeping_test2035.copy()
                    variacao = '20-35'
                    atividade = 'Sleeping'
                if (z==5) & (k==2): #Sleeping 40-55
                    df_test = df_Sleeping_test4055.iloc[:, [3,4,5]].copy()
                    df_trein = df_Sleeping_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_Sleeping_test = df_Sleeping_test4055.copy()
                    variacao = '40-55'
                    atividade = 'Sleeping'
                if (z==5) & (k==3): #Sleeping 60-75
                    df_test = df_Sleeping_test6075.iloc[:, [3,4,5]].copy()
                    df_trein = df_Sleeping_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_Sleeping_test = df_Sleeping_test6075.copy()
                    variacao = '60-75'
                    atividade = 'Sleeping'
                if (z==5) & (k==4): #Sleeping 80-95
                    df_test = df_Sleeping_test8095.iloc[:, [3,4,5]].copy()
                    df_trein = df_Sleeping_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_Sleeping_test = df_Sleeping_test8095.copy()
                    variacao = '80-95'
                    atividade = 'Sleeping'                                   
                if (z==5) & (k==5): #Sleeping 20-95
                    df_test = df_Sleeping_test2095.iloc[:, [3,4,5]].copy()
                    df_trein = df_Sleeping_trein.iloc[0:semana*7].copy()
                    df_trein = df_trein.iloc[:, [3,4,5]].copy()
                    df_Sleeping_test = df_Sleeping_test2095.copy()
                    variacao = '20-95'
                    atividade = 'Sleeping'    
                    

                # standarização
                scaler = StandardScaler().fit(df_trein)
                dfs_trein = scaler.transform(df_trein).copy()
                dfs_test = scaler.transform(df_test).copy()

                #modelo
                # fit the model
                if y == 1:
                    clf = svm.OneClassSVM(nu=0.1, kernel='rbf', gamma='scale')
                    algoritmo = 'ocsvm'
                if y == 2:
                    clf = LocalOutlierFactor(n_neighbors=10, novelty=True)
                    algoritmo = 'lof'
                if y == 3:
                    clf = IsolationForest(contamination= 0.1, random_state=1)
                    algoritmo = 'iforest'
                if y == 4:
                    clf = EllipticEnvelope(contamination= 0.1, random_state=1, support_fraction= 0.9)
                    #clf = EllipticEnvelope(contamination= 0.1, random_state=1, support_fraction= 0.6)
                    algoritmo = 'rce'

                # treinamento e predicao
                clf.fit(dfs_trein)
                y_pred_train = clf.predict(dfs_trein)
                y_pred_test = clf.predict(dfs_test)

                #calculando métricas
                if (z==1): #BedToToilet
                    df_BedToToilet_test['predicao'] = y_pred_test
                    df_BedToToilet_test['vp'] = np.where((df_BedToToilet_test['outlier'] == -1) & (df_BedToToilet_test['predicao'] == -1), 1,0)
                    df_BedToToilet_test['vn'] = np.where((df_BedToToilet_test['outlier'] == 1) & (df_BedToToilet_test['predicao'] == 1), 1,0)
                    df_BedToToilet_test['fp'] = np.where((df_BedToToilet_test['outlier'] == -1) & (df_BedToToilet_test['predicao'] == 1), 1,0)
                    df_BedToToilet_test['fn'] = np.where((df_BedToToilet_test['outlier'] == 1) & (df_BedToToilet_test['predicao'] == -1), 1,0)
                    VP = df_BedToToilet_test['vp'].sum()
                    VN = df_BedToToilet_test['vn'].sum()
                    FP = df_BedToToilet_test['fp'].sum()
                    FN = df_BedToToilet_test['fn'].sum()             

                if (z==2): #Eating
                    df_Eating_test['predicao'] = y_pred_test
                    df_Eating_test['vp'] = np.where((df_Eating_test['outlier'] == -1) & (df_Eating_test['predicao'] == -1), 1,0)
                    df_Eating_test['vn'] = np.where((df_Eating_test['outlier'] == 1) & (df_Eating_test['predicao'] == 1), 1,0)
                    df_Eating_test['fp'] = np.where((df_Eating_test['outlier'] == -1) & (df_Eating_test['predicao'] == 1), 1,0)
                    df_Eating_test['fn'] = np.where((df_Eating_test['outlier'] == 1) & (df_Eating_test['predicao'] == -1), 1,0)
                    VP = df_Eating_test['vp'].sum()
                    VN = df_Eating_test['vn'].sum()
                    FP = df_Eating_test['fp'].sum()
                    FN = df_Eating_test['fn'].sum()   

                if (z==3): #MealPreparation
                    df_MealPreparation_test['predicao'] = y_pred_test
                    df_MealPreparation_test['vp'] = np.where((df_MealPreparation_test['outlier'] == -1) & (df_MealPreparation_test['predicao'] == -1), 1,0)
                    df_MealPreparation_test['vn'] = np.where((df_MealPreparation_test['outlier'] == 1) & (df_MealPreparation_test['predicao'] == 1), 1,0)
                    df_MealPreparation_test['fp'] = np.where((df_MealPreparation_test['outlier'] == -1) & (df_MealPreparation_test['predicao'] == 1), 1,0)
                    df_MealPreparation_test['fn'] = np.where((df_MealPreparation_test['outlier'] == 1) & (df_MealPreparation_test['predicao'] == -1), 1,0)
                    VP = df_MealPreparation_test['vp'].sum()
                    VN = df_MealPreparation_test['vn'].sum()
                    FP = df_MealPreparation_test['fp'].sum()
                    FN = df_MealPreparation_test['fn'].sum()          

                if (z==4): #SpareTimeTV
                    df_SpareTimeTV_test['predicao'] = y_pred_test
                    df_SpareTimeTV_test['vp'] = np.where((df_SpareTimeTV_test['outlier'] == -1) & (df_SpareTimeTV_test['predicao'] == -1), 1,0)
                    df_SpareTimeTV_test['vn'] = np.where((df_SpareTimeTV_test['outlier'] == 1) & (df_SpareTimeTV_test['predicao'] == 1), 1,0)
                    df_SpareTimeTV_test['fp'] = np.where((df_SpareTimeTV_test['outlier'] == -1) & (df_SpareTimeTV_test['predicao'] == 1), 1,0)
                    df_SpareTimeTV_test['fn'] = np.where((df_SpareTimeTV_test['outlier'] == 1) & (df_SpareTimeTV_test['predicao'] == -1), 1,0)
                    VP = df_SpareTimeTV_test['vp'].sum()
                    VN = df_SpareTimeTV_test['vn'].sum()
                    FP = df_SpareTimeTV_test['fp'].sum()
                    FN = df_SpareTimeTV_test['fn'].sum()                   

                if (z==5): #Sleeping
                    df_Sleeping_test['predicao'] = y_pred_test
                    df_Sleeping_test['vp'] = np.where((df_Sleeping_test['outlier'] == -1) & (df_Sleeping_test['predicao'] == -1), 1,0)
                    df_Sleeping_test['vn'] = np.where((df_Sleeping_test['outlier'] == 1) & (df_Sleeping_test['predicao'] == 1), 1,0)
                    df_Sleeping_test['fp'] = np.where((df_Sleeping_test['outlier'] == -1) & (df_Sleeping_test['predicao'] == 1), 1,0)
                    df_Sleeping_test['fn'] = np.where((df_Sleeping_test['outlier'] == 1) & (df_Sleeping_test['predicao'] == -1), 1,0)
                    VP = df_Sleeping_test['vp'].sum()
                    VN = df_Sleeping_test['vn'].sum()
                    FP = df_Sleeping_test['fp'].sum()
                    FN = df_Sleeping_test['fn'].sum()                 

                #calculo das metricas
                erro = (FP+FN)/(VP+VN+FP+FN) #erro
                acuracia =  (VP+VN)/(VP+VN+FP+FN) # accuracy
                precisao = (VP)/(FP+VP) # precision
                sensibilidade = VP/(FN+VP) #recall
                f1escore = 2 * ((precisao * sensibilidade)/(precisao + sensibilidade))
                #armazenando metricas temp
                cvp.append(VP)
                cvn.append(VN)
                cfp.append(FP)
                cfn.append(FN)
                cacuracia.append(acuracia)
                cprecisao.append(precisao)
                csensibilidade.append(sensibilidade)
                cf1escore.append(f1escore)
                
                #gerando dataset metricas
                zipped = list(zip(cvp,cvn,cfp,cfn,cacuracia, cprecisao, csensibilidade, cf1escore))  
                dft_metricas = pd.DataFrame(zipped, columns=['vp', 'vn', 'fp', 'fn','acuracia','precisao','sensibilidade','f1score'])
                dft_metricas['algoritmo'] = algoritmo
                dft_metricas['variacao'] = variacao
                dft_metricas['atividade'] = atividade
                dft_metricas['semana'] = semana
                #dft_metricas.to_csv(r'datasetaruba/results/armetricas-'+atividade+'-'+algoritmo+'-'+variacao+'-sem'+cont_semana+'.csv', sep=';', index=False, decimal=',')
                dfc_metricas = pd.concat([dfc_metricas, dft_metricas.copy()],ignore_index=True)

# print(dfc_metricas.shape)
# dfc_metricas.head(50)            
dfc_metricas.to_csv(r'datasetaruba/results/armetricas-atividades-geral.csv', sep=';', index=False, decimal=',')
```
