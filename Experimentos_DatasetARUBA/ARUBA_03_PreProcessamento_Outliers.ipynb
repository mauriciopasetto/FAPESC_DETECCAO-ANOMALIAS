```python
df_dfreq['BedToToilet'] = df_dur['BedToToilet'] / df_freq['BedToToilet']
df_dfreq['Eating'] = df_dur['Eating'] / df_freq['Eating']
df_dfreq['MealPreparation'] = df_dur['MealPreparation'] / df_freq['MealPreparation']
df_dfreq['Relax'] = df_dur['Relax'] / df_freq['Relax']
df_dfreq['Sleeping'] = df_dur['Sleeping'] / df_freq['Sleeping']

df_dfreq['BedToToilet'] = df_dfreq['BedToToilet'].astype(int)
df_dfreq['Eating'] = df_dfreq['Eating'].astype(int)
df_dfreq['MealPreparation'] = df_dfreq['MealPreparation'].astype(int)
df_dfreq['Relax'] = df_dfreq['Relax'].astype(int)
df_dfreq['Sleeping'] = df_dfreq['Sleeping'].astype(int)

df_dfreq.head()
```
