```python
#LOF
#ACURACIA
sns.set_theme(font_scale=1.3)
ax = sns.lineplot(x="semanas", y="acuracia", data=df_lof_2035, color='red')
ax = sns.lineplot(x="semanas", y="acuracia", data=df_lof_4055, color='green')
ax = sns.lineplot(x="semanas", y="acuracia", data=df_lof_6075, color='blue')
ax = sns.lineplot(x="semanas", y="acuracia", data=df_lof_8095, color='magenta')
ax = sns.lineplot(x="semanas", y="acuracia", data=df_lof_2095, color='brown')

ax.set_xlabel("Semanas")
ax.set_ylabel("Acurácia")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.savefig('all_lof_att.svg')
```
